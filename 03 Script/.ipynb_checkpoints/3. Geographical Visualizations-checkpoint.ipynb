```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import os
import folium
import json
import geopandas as gpd
```

```python
# This option ensures the charts I create are displayed in the notebook without the need to "call" them specifically.
%matplotlib inline
```

```python
# Import ".json" file for the U.S. City

city_geo = r'C:\Users\vicky\OneDrive\文件\careerfoundry\Data Analytics Immersion\Exercise 6\05-2024 Project\02 Data\California_Cities.geojson'
```

```python
# Define path
path = r'C:\Users\vicky\OneDrive\文件\careerfoundry\Data Analytics Immersion\Exercise 6\05-2024 Project'
```

```python
# import data & only include the variables columns that make sense for exploratory visual analysis
vars_list = ["Customer ID", "City", "Zip Code", "Satisfaction Score"]
```

```python
df = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Telco_cleaned.csv'), usecols = vars_list)
```

```python
df.head()
```

```python
# Create a data frame with just the city and the values for rating we want plotted
data_to_plot = df[['City','Satisfaction Score']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [36.7783, -119.4179], zoom_start = 6)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = city_geo, 
    data = data_to_plot,
    columns = ['City','Satisfaction Score'],
    key_on = 'feature.properties.FID', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "rating").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
city_geo3 = r'C:\Users\vicky\Downloads\CaliforniaCities\CalifCities.geojson'
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [36.7783, -119.4179], zoom_start = 6)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = city_geo3, 
    data = data_to_plot,
    columns = ['ObjectID','Satisfaction Score'],
    key_on = 'feature.properties.ObjectID', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "Satisfaction Score").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
# Create a Choropleth map
folium.Choropleth(
    geo_data=city_geo3,
    name='choropleth',
    data=data_to_plot,
    columns=['City', 'Satisfaction Score'],
    key_on='feature.properties.ObjectID',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Satisfaction Score'
).add_to(map)
```
